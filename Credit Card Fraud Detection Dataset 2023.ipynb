### Credit Card Fraud Detection Dataset 2023 

LINK: https://www.kaggle.com/datasets/nelgiriyewithana/credit-card-fraud-detection-dataset-2023

![](https://storage.googleapis.com/kaggle-datasets-images/3752264/6492730/3a84e7d4591f3f4323eebcd0a4ec4114/dataset-cover.jpg?t=2023-09-18-10-01-48)

```python
import kagglehub
import pandas as pd
import os
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline

# Download latest version
path = kagglehub.dataset_download("nelgiriyewithana/credit-card-fraud-detection-dataset-2023")

print("Path to dataset files:", path)

# Find the CSV file in the download path
csv_file = None
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            csv_file = os.path.join(root, file)
            print("CSV file found:", csv_file)
            break

# Check if the CSV file was found
if csv_file:
    # Convert to DataFrame
    df = pd.read_csv(csv_file)
    print("CSV file found in the dataset directory.")
else:
    print("CSV file not found in the dataset directory.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizar a distribuição de transações fraudulentas vs não fraudulentas
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Class', data=df)
plt.title('Distribuição de Transações Fraudulentas vs Não Fraudulentas')
plt.xlabel('Classe')
plt.ylabel('Número de Transações')

# Adicionar rótulos nos valores das barras
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

Contagem de transações por classe

```python
# Contagem de transações por classe
class_counts = df['Class'].value_counts()
print(class_counts)
```

Verificar a Consistência dos Dados:

Se ambos os valores forem realmente iguais, podemos fazer uma verificação adicional para garantir que os dados estejam corretos e não tenham sido duplicados ou inseridos erroneamente.

```python
# Verificar duplicatas
duplicate_rows = df.duplicated().sum()
print(f"Número de linhas duplicadas: {duplicate_rows}")

# Verificar se há uma possível anomalia ou erro na coluna 'Class'
if class_counts[0] == class_counts[1]:
    print("Os valores para transações fraudulentas e não fraudulentas são idênticos.")
else:
    print("Há uma diferença nos valores entre transações fraudulentas e não fraudulentas.")
```

Visto que o número de transações fraudulentas e não fraudulentas é idêntico e não há duplicatas, podemos prosseguir com a análise sabendo que os dados são balanceados e confiáveis.

Preparação dos Dados para Modelagem
Vamos preparar os dados para a criação de modelos de detecção de fraude:

```python
from sklearn.model_selection import train_test_split

# Selecionar variáveis preditoras (V1-V28 e Amount) e a variável alvo (Class)
X = df.drop(columns=['id', 'Class'])
y = df['Class']

# Dividir o conjunto de dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Criação de Modelos de Detecção de Fraude

#### Vamos começar com um modelo de Árvore de Decisão:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Treinar o modelo de árvore de decisão
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação:\n", classification_report(y_test, y_pred))

# Matriz de Confusão
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:\n", conf_matrix)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Treinar o modelo de Floresta Aleatória
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_rf = rf_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Random Forest:\n", classification_report(y_test, y_pred_rf))

# Matriz de Confusão
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Matriz de Confusão - Random Forest:\n", conf_matrix_rf)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_rf.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Random Forest')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Treinar o modelo de Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_gb = gb_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Gradient Boosting:\n", classification_report(y_test, y_pred_gb))

# Matriz de Confusão
conf_matrix_gb = confusion_matrix(y_test, y_pred_gb)
print("Matriz de Confusão - Gradient Boosting:\n", conf_matrix_gb)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_gb.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_gb, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Gradient Boosting')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

# Treinar o modelo de Redes Neurais
nn_model = MLPClassifier(random_state=42, max_iter=300)
nn_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_nn = nn_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Redes Neurais:\n", classification_report(y_test, y_pred_nn))

# Matriz de Confusão
conf_matrix_nn = confusion_matrix(y_test, y_pred_nn)
print("Matriz de Confusão - Redes Neurais:\n", conf_matrix_nn)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_nn.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Redes Neurais')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

### Criando Novas Features no DataFrame

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Criar algumas novas features para exemplo
df['V1_V2'] = df['V1'] * df['V2']
df['log_Amount'] = np.log1p(df['Amount'])

# Atualizar variáveis preditoras
X = df.drop(columns=['id', 'Class'])
y = df['Class']

# Dividir o conjunto de dados em treino e teste novamente
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Treinar o modelo de Floresta Aleatória
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_rf = rf_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Random Forest:\n", classification_report(y_test, y_pred_rf))

# Matriz de Confusão
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Matriz de Confusão - Random Forest:\n", conf_matrix_rf)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_rf.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Random Forest')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Treinar o modelo de Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_gb = gb_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Gradient Boosting:\n", classification_report(y_test, y_pred_gb))

# Matriz de Confusão
conf_matrix_gb = confusion_matrix(y_test, y_pred_gb)
print("Matriz de Confusão - Gradient Boosting:\n", conf_matrix_gb)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_gb.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_gb, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Gradient Boosting')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

# Treinar o modelo de Redes Neurais
nn_model = MLPClassifier(random_state=42, max_iter=300)
nn_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_nn = nn_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Redes Neurais:\n", classification_report(y_test, y_pred_nn))

# Matriz de Confusão
conf_matrix_nn = confusion_matrix(y_test, y_pred_nn)
print("Matriz de Confusão - Redes Neurais:\n", conf_matrix_nn)

# Matriz de Confusão em números
tn, fp, fn, tp = conf_matrix_nn.ravel()
print(f"Verdadeiros Negativos (TN): {tn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"Falsos Negativos (FN): {fn}")
print(f"Verdadeiros Positivos (TP): {tp}")

# Visualização da Matriz de Confusão
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Redes Neurais')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```

## Conclusão

Random Forest ainda apresenta a melhor performance entre os três modelos, com precisão, recall e F1-score perfeitos. A matriz de confusão mostra pouquíssimos erros, com apenas 22 falsos positivos e 0 falsos negativos.

Gradient Boosting e Redes Neurais apresentam uma performance inferior em comparação ao Random Forest, com valores de precisão e recall ligeiramente menores e um número maior de erros (FP e FN).

Insight Geral: A introdução das novas features parece ter beneficiado significativamente o modelo Random Forest, mantendo um desempenho quase perfeito. Gradient Boosting e Redes Neurais também se beneficiaram das novas features, mas ainda ficam aquém do desempenho da Random Forest.

```python
from sklearn.model_selection import cross_val_score

# Validar Random Forest com Cross-Validation
rf_scores = cross_val_score(rf_model, X, y, cv=5)
print(f"Cross-Validation Scores - Random Forest: {rf_scores}")
print(f"Média dos Scores: {rf_scores.mean()}")

# Validar Gradient Boosting com Cross-Validation
gb_scores = cross_val_score(gb_model, X, y, cv=5)
print(f"Cross-Validation Scores - Gradient Boosting: {gb_scores}")
print(f"Média dos Scores: {gb_scores.mean()}")

# Validar Redes Neurais com Cross-Validation
nn_scores = cross_val_score(nn_model, X, y, cv=5)
print(f"Cross-Validation Scores - Redes Neurais: {nn_scores}")
print(f"Média dos Scores: {nn_scores.mean()}")
```

## Conclusão:
A criação das novas features teve um impacto positivo notável no modelo de Gradient Boosting, melhorando sua média de scores.

Para Random Forest, a mudança foi mínima, com uma leve queda, mas a performance ainda permanece excelente.

No caso das Redes Neurais, houve uma pequena queda nos scores, mas a performance ainda é bastante alta.

## Remover as Novas Features e Retrainar o Modelo

Como a mudança para Random Forest (Teve melhores resultados) foi mínima e teve até uma leve queda, vamos remover as novas features e retreinar o modelo Random Forest. Depois vamos salvar o modelo e usar posteriormente para realizar previsões em um conjunto novo conjunto de dados.

```python
# Remover as colunas das novas features
df = df.drop(columns=['V1_V2', 'log_Amount'])

# Atualizar variáveis preditoras
X = df.drop(columns=['id', 'Class'])
y = df['Class']

# Dividir o conjunto de dados em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Treinar o modelo de Floresta Aleatória
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred_rf = rf_model.predict(X_test)

# Avaliar o modelo
print("Relatório de Classificação - Random Forest:\n", classification_report(y_test, y_pred_rf))

# Matriz de Confusão
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Matriz de Confusão - Random Forest:\n", conf_matrix_rf)

# Visualização da Matriz de Confusão
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão - Random Forest')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()
```
