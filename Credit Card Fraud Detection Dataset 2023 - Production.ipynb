```python
import numpy as np
import pandas as pd

# Carregar o novo conjunto de dados
df = pd.read_csv('Credit_Card_Fraud_Detection_Dataset_2023_New.csv')

# Verifique se o novo conjunto de dados está carregado corretamente
df.head(5)
```

```python
# Criar um novo índice numérico
df['id'] = range(len(df))
```

```python
# Verificar as primeiras linhas para confirmar a mudança
df.head(20)
```

```python
import joblib

# Carregar o modelo treinado
rf_model = joblib.load('random_forest_model_final.pkl')
print("Modelo Random Forest carregado com sucesso!")
```

```python
# Remover colunas que não são features
X = df.drop(columns=['id', 'Class'])

# Verificar o DataFrame final de features
X.head()
```

```python
# Fazer previsões no novo conjunto de dados
predicoes = rf_model.predict(X)

# Adicionar previsões ao DataFrame original
df['Previsao'] = predicoes
print("Previsões adicionadas ao DataFrame:\n", df[['id', 'Previsao']])
```

```python
from sklearn.metrics import classification_report

# Relatório de Classificação
print("Relatório de Classificação:\n", classification_report(df['Class'], df['Previsao']))
```

### Como estamos lidando com transações bancarias, vamos tentar usar tecnicas para melhorar os resultados ao máximo. (OPCIONAL)

#### Ajuste de Hiperparâmetros usando GridSearchCV:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Carregar o modelo treinado salvo
import joblib
rf_model = joblib.load('random_forest_model_final.pkl')

# Definir o espaço de hiperparâmetros
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Inicializar o GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Separar features e labels
X = df.drop(columns=['id', 'Class'])
y = df['Class']

# Realizar o ajuste de hiperparâmetros
grid_search.fit(X, y)

# Imprimir os melhores hiperparâmetros e a melhor acurácia
print("Melhores Hiperparâmetros: ", grid_search.best_params_)
print("Melhor Acurácia: ", grid_search.best_score_)
```

#### Balanceamento de Dados usando SMOTE:

```python
from imblearn.over_sampling import SMOTE

# Inicializar SMOTE
smote = SMOTE(random_state=42)

# Aplicar SMOTE para balancear os dados
X_resampled, y_resampled = smote.fit_resample(X, y)

# Verificar o novo tamanho do conjunto de dados balanceado
print("Tamanho original:", X.shape)
print("Tamanho balanceado:", X_resampled.shape)
```

#### Treinar o Modelo com Dados Balanceados e Hiperparâmetros Ajustados:

```python
# Treinar o modelo com os dados balanceados e hiperparâmetros ajustados
rf_model_balanced = RandomForestClassifier(
    n_estimators=grid_search.best_params_['n_estimators'],
    max_features=grid_search.best_params_['max_features'],
    max_depth=grid_search.best_params_['max_depth'],
    min_samples_split=grid_search.best_params_['min_samples_split'],
    min_samples_leaf=grid_search.best_params_['min_samples_leaf'],
    random_state=42
)

rf_model_balanced.fit(X_resampled, y_resampled)

# Fazer previsões com o modelo balanceado
y_pred_balanced = rf_model_balanced.predict(X)

# Avaliar o modelo balanceado
from sklearn.metrics import classification_report

print("Relatório de Classificação com Dados Balanceados:\n", classification_report(y, y_pred_balanced))
```

#### Avaliar o Desempenho do Modelo:

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# Matriz de Confusão
conf_matrix = confusion_matrix(y, y_pred_balanced)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Não Fraudulenta', 'Fraudulenta'], yticklabels=['Não Fraudulenta', 'Fraudulenta'])
plt.title('Matriz de Confusão')
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.show()

# Acurácia
acuracia = accuracy_score(y, y_pred_balanced)
print(f'Acurácia: {acuracia * 100:.2f}%')

# Curva ROC e AUC
auc_score = roc_auc_score(y, y_pred_balanced)
print(f'AUC: {auc_score:.2f}')
fpr, tpr, thresholds = roc_curve(y, y_pred_balanced)
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='Curva ROC')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```
